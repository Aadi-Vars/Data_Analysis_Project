```python
def multi_choice(series):
    result=series.to_frame()
    options=[]

    for idx,value in series[series.notnull()].items():
        for option in value.split(";"):
            if not option in result.columns:
                options.append(option)
                result[option]=False
            result.at[idx,option]=True
    return result[options]
```

```python
stackoverflow_2022_df_employment_data=multi_choice(stackoverflow_2022_df.Employment)
stackoverflow_2022_df_employment_data
```

```python
stackoverflow_2023_df_employment_data=multi_choice(stackoverflow_2023_df.Employment)
stackoverflow_2023_df_employment_data
```

```python

employment_final_2022=stackoverflow_2022_df_employment_data.sum().sort_values()
employment_final_2023=stackoverflow_2023_df_employment_data.sum().sort_values()
```

```python
fig,axes=plt.subplots(2,1)
fig.tight_layout(pad=2)
axes[0].barh(employment_final_2022.index,employment_final_2022);
axes[0].set_title(scheme_2022_df.Employment.split("?")[0]+"(2022)");

axes[1].barh(employment_final_2023.index,employment_final_2023);
axes[1].set_title(scheme_2023_df.Employment.split("?")[0]+"(2023)");
```
